```python
import pandas as pd
import numpy as np
import os
```

```python
regions = pd.read_csv("/data/KNW/f.tijhuis/Atlases_CIFTI/Glasser/Cortical+Freesurfer/GlasserFreesurfer_region_names_full.txt",
                     header=None)
names = pd.read_csv("/data/KNW/f.tijhuis/Atlases_CIFTI/Glasser/Cortical+Freesurfer/GlasserFreesurfer_subnet_order_names.txt",
                     header=None)
```

```python
subnetworks = pd.DataFrame()
subnetworks["region"] = list(regions[0])
subnetworks["subnetwork"] = list(names[0])
```

```python
FPN = list(subnetworks[subnetworks['subnetwork']=='FP'].index.values)
DMN = list(subnetworks[subnetworks['subnetwork']=='DMN'].index.values)
FPN_names = list(subnetworks['region'][FPN])
DMN_names = list(subnetworks['region'][DMN])

all_node_names = list(subnetworks['region'])
```

```python
len(FPN)
```

```python
FPN_L = [i for i in FPN_names if '_L' in i]
FPN_R = [i for i in FPN_names if '_R' in i]
```

```python
FPN_0 = [i[:-2] for i in FPN_L]
FPN_1 = [i[:-2] for i in FPN_R]
FPN_2 = FPN_0 + FPN_1
FPN_2 = set(FPN_2)
```

```python
fpn = pd.DataFrame()
fpn['0'] = list(FPN_2)
```

```python
combined_R = '\t'.join(FPN_R)
combined_L = '\t'.join(FPN_L)
```

```python
L_list = []
for i in fpn['0']:
    
    if i in combined_L:   
        L_list.append(i)
    else:
        L_list.append('-')
fpn['L'] = L_list
R_list = []

for i in fpn['0']:
    
    if i in combined_R:   
        R_list.append(i)
    else:
        R_list.append('-')
fpn['R'] = R_list
```
