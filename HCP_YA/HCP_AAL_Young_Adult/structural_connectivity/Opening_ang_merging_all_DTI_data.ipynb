```python
import pandas as pd
```

```python
import glob
```

```python
files=glob.glob('*connectivity*txt')
```

```python
files
```

```python
import numpy as np
Data=[]
for file in files:
    DTI=pd.read_csv(file,sep='\t',header=1)
    DTI=DTI.set_index('data.1')
    DTI=DTI.drop(['data', 'Unnamed: 118'], axis=1)
    DTI /=np.max(np.abs(DTI))
    Data.append(DTI)
    
```

```python
import pickle
open_file = open('DTI.txt', "wb")
pickle.dump(Data, open_file)
open_file.close()

open_file = open('DTI.txt', "rb")
loaded_DTI = pickle.load(open_file)
open_file.close()
```
